# Was ist Python?

- Python ist eine Programmiersprache, die in verschiedenen Bereichen eingesetzt wird (Softwareentwicklung, Webentwicklung, Datenanalyse).
- Entwickelt von Guido van Rossum, der von der Community genannt „lebenslanger wohlwollender Diktator“, bis er im Jahr 2018 von seinem „Amt“ zurücktrat. 📘[Link](https://news.microsoft.com/de-de/developer-stories-guido-van-rossum/)
- Python ist eine so genannte „höhere“ Programmiersprache, d. h. sie ist weiter von der Maschinenebene entfernt, aber die Syntax ist für Menschen leichter zu verstehen. 
- Ein „Interpreter“ wandelt Code von einem für Menschen verständlichen Code in Maschinencode um. 📘[Link](https://de.wikipedia.org/wiki/Interpreter)
- Zahlreiche Bibliotheken erweitern die Grundfunktionalität von Python in verschiedenen Bereichen. Python-Pakete können zum Beispiel von PyPI (Python Package Index) oder über Anaconda heruntergeladen werden.
- Laut den auf GitHub gesammelten Daten ist Python eine der am häufigsten genutzten Programmiersprachen der Welt. 📘[Link](https://madnight.github.io/githut/#/stars/2023/1)

# Was versteht man unter Data Analytics?

- Der Begriff Data Analytics umfasst alle Aktivitäten zur Gewinnung von Informationen aus Daten. 
- Das sind das Sammeln, Speichern, Organisieren der Daten sowie die Analyse der Daten und die Präsentation der Ergebnisse. 📘[Link](https://www.computerwoche.de/a/was-ist-data-analytics,3550603)
- Manchmal werden die Begriffe Data Analytics und Data Analysis (auf deutsch Datenanalyse) als synonym verwendet. Es gibt jedoch einen kleinen Unterschied. 
- Data Analytics ist ein umfassenderes Konzept. Das bezieht sich auf den Prozess der Verwendung von Daten um Erkenntnisse zu gewinnen und fundierte Entscheidungen zu treffen. 📘[Link](https://www.bmc.com/blogs/data-analytics-vs-data-analysis/)
- Die Datenanalyse ist ein fundamentaler Teil der Data Analytics. Sie bezieht sich auf das Bereinigen, Umwandeln und Modellieren von Daten, um nützliche Informationen zu finden.

# Die vier Methoden der Data Analytics

Die Data Analytics wird zusätlich in Aktivitäten bzw Methoden unterteilt:
 - Descriptive Analytics - Was ist passiert?
 - Diagnostic Analytics - Warum ist das passiert?
 - Predictive Analytics - Was wird in der Zukunft passieren?
 - Prescriptive Analytics - Was sollte getan werden, damit dies passiert?

# Data Analysis vs Data Science

**Data Analysis:**
- Der Datenanalyst befasst sich mit der Sammlung und Aufbereitung von Daten sowie der Untersuchung von Trends, Beziehungen und einfachen Prognosemodellen.
- Zu den Instrumenten der Datenanalyse gehören hauptsächlich deskriptive Statistiken, Regressionen (linear und logit), Zeitreihenökonometrie und Clusteranalyse.
- Andere Skills sind sehr wichtig: Präsentationen halten, Berichte schreiben.

**Data Science:**
- Data Science, auf deutsch Datenwissenschaft, konzentriert sich hauptsächlich auf die Predictive Analytics.
- In Data Science werden fortgeschrittene Tools genutz, wie Machine Learning, Deep Learning und natural language processing (zur Analyse von Texten).
- Mehr Kenntnisse in Mathematik und Statistik sind erforderlich.
- Der Data Scientist kümmert sich um das Modell und die Aktualisierung der Modellparameter, wenn neue Daten zur Verfügung stehen.
- Data Science $\neq$ Data Engineering.
